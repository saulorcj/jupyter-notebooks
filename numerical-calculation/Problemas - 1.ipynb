```python
import matplotlib.pyplot as plt
```

```python
a = 1.0
c = 0.75
α, γ = 0.5, 0.5
```

```python
def dev_x(x, y):
    return a * x - α * x * y
```

```python
def dev_y(x, y):
    return -c * y + γ * x * y
```

```python
def met_euler(x, y, h, limite):
    valores_x, valores_y = [], []
    t = 0
    while t <= limite:
        y_ = y + h * dev_y(x, y)
        x_ = x + h * dev_x(x, y)
        if x_ < 0:
            x_ = 0
        if y_ < 0:
            y_ = 0
        valores_x.append(x_)
        valores_y.append(y_)
        x = x_
        y = y_
        t += h
    return valores_x, valores_y
```

```python
def get_grid(start, end, jump):
    values = []
    h = start
    while h <= end:
        values.append(h)
        h += jump
    return values
```

```python
def plotar(x, y, h, limite=100):
    # Realizando plot
    # Convertendo respostas para numpy
    x_num, y_num = 1, 1
    #beta_num = np.array(solution).astype(float).flatten()

    # Gerando a malha (grid)
    x_min, x_max = 0, limite
    # Calculando o valor da função (no caso, a reta) nesta malha (grid) gerada
    t_grid = get_grid(x_min, x_max, h)
    x_grid, y_grid = met_euler(x, y, h, limite)
    print(x_grid[-1], y_grid[-1])
    """for i in range(len(t_grid)):
        print(t_grid[i], x_grid[i], y_grid[i])"""
    # Instruções do plot
    fig, ax = plt.subplots(figsize = (16,8))
    ax.plot(t_grid, x_grid, linewidth=3) # reta
    ax.plot(t_grid, y_grid, linewidth=3) # reta
    ax.plot(x_num, y_num, linestyle = "None") # pontos do conjunto de dados

    # Labels nos eixos, limites do plot e legenda
    ax.set_xlabel("$t$", size = 16)
    ax.set_ylabel("$p$", size = 18)
    ax.set_xlim(x_min, x_max)
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.legend(["$x(t)$", "$y(t)$"], loc = "upper left", fontsize = 15)
    ax.grid(True)

    plt.show()
```

```python
x = 2
y = 5
h = 0.0001

plotar(x, y, h)
```
